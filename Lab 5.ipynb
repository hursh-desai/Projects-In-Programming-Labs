```python
from bs4 import BeautifulSoup
import requests
```

Lab 5.1

```python
url = "http://www.imdb.com/search/title?sort=num_votes,desc&start=1&title_type=feature&year=1900,2015"
page = requests.get(url)
bs = BeautifulSoup(page.text, 'html.parser')
movies = bs.findAll('div','lister-item-content')
```

```python
movie_1 = movies[0]
```

```python
description = movie_1.findAll('p')[1].contents[0]
description.strip()
```

```python
for movie in movies:
    title = movie.find('a').contents[0]
    genre = movie.find('span', 'genre').contents[0]
    genre = genre.lstrip("\n")
    runtime = movie.find('span','runtime').contents[0]
    rating = movie.find('span','value').contents[0]
    year = movie.find('span','lister-item-year text-muted unbold').contents[0]
    mpaa_rating = movie.find('span','certificate').contents[0]
    description = movie.findAll('p')[1].contents[0]
    print ('Title: {}\nGenre: {}\nRuntime: {}\nMPAA Rating: {}\nRating: {}\nYear: {}\nDescription: {}'.format
           (title, genre, runtime, mpaa_rating, rating, year, description.strip()))
    print('-----------------------------------------------------------------------------------')
```

```python
url = "https://www.basketball-reference.com/teams/TOR/2019_games.html#games_link"
page = requests.get(url)
bs = BeautifulSoup(page.text, 'html.parser')
body = bs.find('tbody')
games_pts = body.findAll('td', {'class':"right", 'data-stat': "pts"})
games_opp_pts = body.findAll('td', {'class':"center", 'data-stat': "opp_pts"})
points = [int(game.contents[0]) for game in games_pts]
opp_points = [int(game.contents[0]) for game in games_opp_pts]
differential = [a - b for a, b in zip(points, opp_points)]
```

```python
url = "https://www.basketball-reference.com/players/s/siakapa01/gamelog/2019/"
page = requests.get(url)
bs = BeautifulSoup(page.text, 'html.parser')
body = bs.find('tbody')
games_mts = body.findAll('td', {'class':"right", 'data-stat': "mp"})
minutes = [game.contents[0] for game in games_mts]
minutes[31:31] = ['0']
minutes[76:76] = ['0']
ftr = [60,1]
seconds = [sum([a*b for a,b in zip(ftr, map(int,minute.split(':')))]) for minute in minutes]
```

```python
#basic code to connect to a database. 
from sqlalchemy import create_engine

# Must have own parameters for MySQL database

conn_string = 'mysql://{user}:{password}@{host}/{db}?charset={encoding}'.format(
    host = ' ', 
    user = ' ',
    db = ' ',
    password = ' ',
    encoding = 'utf8mb4')

engine = create_engine(conn_string)
con = engine.connect()
print(con)
```

```python
db_name = 'basketball'
create_db_query = "CREATE DATABASE IF NOT EXISTS {db} DEFAULT CHARACTER SET 'utf8'".format(db=db_name)
con.execute(create_db_query)
print(db_name)
```

```python
db = "basketball"
table_name = 'siakam_min_vs_diff'
create_table_query = '''CREATE TABLE IF NOT EXISTS {db}.{table} 
                                (game int NOT NULL AUTO_INCREMENT,
                                seconds int,
                                differential int,
                                PRIMARY KEY(game)
                                )'''.format(db=db_name, table=table_name)
con.execute(create_table_query)
```

```python
table_name = 'siakam_min_vs_diff'
query_template = '''INSERT IGNORE INTO {db}.{table}(seconds, differential) 
                    VALUES (%s, %s)'''.format(db=db_name, table=table_name)

for i in range(len(seconds)):
    seconds_i = seconds[i]
    differential_i = differential[i]
                       
    print("Inserting game", i)
    query_parameters = (seconds_i, differential_i)
    con.execute(query_template, query_parameters)
```
