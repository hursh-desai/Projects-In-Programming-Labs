```python
import requests
url = 'http://www.stern.nyu.edu/faculty/search_name_form'
response = requests.get(url)
html = response.text
html
```

```python
type(html)
```

```python
import re

emails = []
regex = re.compile(r""":\S+@stern.nyu.edu\"""")

matches = regex.finditer(html)
for match in matches:
    print(match.group())
    emails.append(match.group())
#     print("Formatted:", match.group(1),"-", match.group(2), "-", match.group(3))
    # print("Starts at:", match.start())
    # print("Ends at:", match.end())
    print("===========")
```

```python
emails1 = []
for email in emails:
    email = email.replace(':', '')
    email = email.replace('"', '')
    emails1.append(email) 
```
