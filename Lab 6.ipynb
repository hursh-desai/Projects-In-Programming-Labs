```python
tv = pd.read_csv('tv.csv')
```

```python
tv
```

```python
tv.plot(kind='scatter', x='av_rating', y='seasonNumber')
```

```python
tv.loc[tv['seasonNumber'] >= 7].plot(kind='scatter', x='av_rating', y='seasonNumber')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(tv.loc[tv['seasonNumber'] >= 7]["av_rating"], tv.loc[tv['seasonNumber'] >= 7]["seasonNumber"])
```

```python
svu = tv.loc[tv['title'] == 'Law & Order: Special Victims Unit', 'av_rating']
```

```python
svu.plot(kind='line', title='Historical Ratings', color='green')
```

```python
tv.loc[(tv['date'] == 2015) & (tv['genres'] == 'Crime,Drama,Mystery')].sort_values(by='av_rating', ascending=False)
```

```python
tv.loc[(tv['date'] == 2015) & (tv['genres'] == 'Crime,Drama,Mystery'), 'av_rating'].mean()
```

Law & Order: Special Victims Unit ranked second to last on all seasons of shows that are in the genre Crime,Drama,Mystery in the year 2015. it was about 1.4 points below the average for all the seasons of shows in that genre for 2015.
